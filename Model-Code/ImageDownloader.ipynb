```python
import pandas as pd
import os
from bing_image_downloader import downloader
csv_file_path = "/content/language-of-flowers.csv"
flowers_df = pd.read_csv(csv_file_path)

flowers_df['Color'] = flowers_df['Color'].fillna("NaN")
root_folder = "Flower_Images"

def download_flower_images(dataframe, root_folder, num_images=1):
    os.makedirs(root_folder, exist_ok=True)

    for _, row in dataframe.iterrows():
        flower = row['Flower']
        color = row['Color']
        search_term = flower if color == "NaN" else f"{flower} {color}"
        file_name = f"{flower}_{color.lower()}.jpg" if color != "NaN" else f"{flower}.jpg"
        print(f"Downloading {num_images} image(s) for: {search_term}")
        downloader.download(
            search_term,
            limit=num_images,
            output_dir=root_folder,
            adult_filter_off=False,
            force_replace=False,
            timeout=60
        )
        downloaded_folder = os.path.join(root_folder, search_term)
        if os.path.exists(downloaded_folder):
            images = [img for img in os.listdir(downloaded_folder) if img.lower().endswith((".jpg", ".jpeg", ".png"))]
            if images:
                os.rename(
                    os.path.join(downloaded_folder, images[0]),
                    os.path.join(root_folder, file_name)
                )
            os.rmdir(downloaded_folder)

download_flower_images(flowers_df, root_folder, num_images=1)

print("Image download and organization completed successfully!")
```

```python
import shutil
import os

def zip_folder(folder_path, output_zip_path):

    shutil.make_archive(output_zip_path, 'zip', folder_path)
    print(f"Folder zipped successfully: {output_zip_path}.zip")


folder_to_zip = "Flower_Images"
output_zip_file = "2Flower_Images_Zip"
zip_folder(folder_to_zip, output_zip_file)
print(f"Download the zip file: {output_zip_file}.zip")
```
