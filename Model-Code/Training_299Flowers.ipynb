```python
import os
import random
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
from sklearn.metrics import accuracy_score
from tqdm.notebook import tqdm
import timm
```

```python
class FlowerDataset(Dataset):
    def __init__(self, data_prep, transform=None):
        self.data_prep = data_prep
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
        ])
```

```python
class FlowerDataset(Dataset):
    def __init__(self, data_prep, transform=None):
        self.data_prep = data_prep
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
        ])

    def __getitem__(self, idx):
        try:
            path, label = self.data_prep[idx]
            image = Image.open(path).convert("RGB")
            if self.transform:
                image = self.transform(image)
            return True, image, torch.tensor(label).long()
        except:
            return False, torch.randn(3, 224, 224), torch.tensor(-1).long()

    def __len__(self):
        return len(self.data_prep)

```

```python
for param in model.parameters():
    param.requires_grad = True

# Define loss and optimizer
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=0.003)
```

```python
def train(model, criterion, optimizer, train_dataloader, test_dataloader, num_epoch=15, path="./", score_function=accuracy_score):
    val_acc_log = []

    for epoch in tqdm(range(num_epoch)):
        model.train()
        pbar = tqdm(train_dataloader)
        for flag, imgs, labels in pbar:
            imgs = torch.stack([imgs[i] for i in range(len(flag)) if flag[i]])
            labels = torch.stack([labels[i] for i in range(len(flag)) if flag[i]])
            if len(labels):
                optimizer.zero_grad()
                imgs = imgs.cuda()
                labels = labels.cuda()
                y_pred = model(imgs)
                loss = criterion(y_pred, labels)
                pbar.set_description(f"Loss: {loss.item():.6f}")
                loss.backward()
                optimizer.step()


        model_path = os.path.join(path, f"ConvNeXt_epoch_{epoch + 1}.pth")
        torch.save(model.state_dict(), model_path)
        print(f"Model saved to: {model_path}")


        val_pred = []
        val_true = []
        model.eval()
        with torch.no_grad():
            for flag, imgs, labels in test_dataloader:
                imgs = torch.stack([imgs[i] for i in range(len(flag)) if flag[i]])
                labels = torch.stack([labels[i] for i in range(len(flag)) if flag[i]])
                if len(labels):
                    imgs = imgs.cuda()
                    labels = labels.cuda()
                    pred = model(imgs)
                    val_pred.extend(pred.argmax(1).cpu().numpy())
                    val_true.extend(labels.cpu().numpy())

        val_acc = score_function(val_true, val_pred)
        val_acc_log.append(val_acc)
        print(f"Epoch {epoch + 1}, Validation Accuracy: {val_acc:.4f}")

    return val_acc_log
```

```python
from torchvision import transforms
from PIL import Image
import os

roses_folder = os.path.join(dataset_path, "Roses")

# Define the same transform as used in training
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
])


test_images = [os.path.join(roses_folder, file) for file in os.listdir(roses_folder) if file.endswith(("jpg", "jpeg", "png"))]
print(f"Total test images: {len(test_images)}")
```

```python
import os
from tqdm.notebook import tqdm

selected_classes = [
    "amaryllis", "anemone", "aster", "begonia", "black-eyedsusan", "borage",
    "camellia", "candytuft", "carnation", "chrysanthemum", "clematis", "clover",
    "columbine", "coreopsis", "crocus", "cyclamen", "daffodil", "dahlia", "daisy",
    "dill", "forgetmenot", "gardenia", "geranium", "gladiolas", "goldenrod",
    "heliotrope", "hibiscus", "holly", "hollyhock", "honeysuckle", "hyacinth",
    "hydrangea", "hyssop", "iris", "jasmine", "larkspur", "lavender", "lilac",
    "callalily", "daylily", "lilyofthevalley", "magnolia", "marigold",
    "morningglory", "nasturtium", "pansy", "peony", "poppy", "roses", "sage",
    "snapdragon", "speedwell", "sunflower", "sweetpea", "tulip", "valerian",
    "viola", "wallflower", "yarrow", "zinnia"
]

dataset_folders = {folder.lower(): folder for folder in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, folder))}
selected_folders = {cls: dataset_folders[cls] for cls in selected_classes if cls in dataset_folders}


label_mapper = {class_name: idx for idx, class_name in enumerate(selected_folders.keys())}


train_data = []

for class_name, label in tqdm(label_mapper.items()):
    folder_path = os.path.join(dataset_path, selected_folders[class_name])
    images = [os.path.join(folder_path, file_name) for file_name in os.listdir(folder_path) if file_name.endswith(("jpg", "jpeg", "png"))]


    train_data.extend([[img, label] for img in images])

print(f"Classes being trained: {len(label_mapper)}")
print(f"Total training images: {len(train_data)}")
```

```python
from torch.utils.data import DataLoader


train_dataset = FlowerDataset(train_data)

train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True, num_workers=2, pin_memory=True)

print(f"Train dataset size: {len(train_dataset)}")
```

```python
import timm
import torch


model = timm.create_model('convnext_tiny', pretrained=True, num_classes=len(label_mapper), in_chans=3).cuda()

for param in model.parameters():
    param.requires_grad = True


criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=0.003)

def train_full_data(model, criterion, optimizer, train_dataloader, num_epoch=20, path="./"):
    for epoch in tqdm(range(num_epoch)):
        model.train()
        pbar = tqdm(train_dataloader)
        for flag, imgs, labels in pbar:
            imgs = torch.stack([imgs[i] for i in range(len(flag)) if flag[i]])
            labels = torch.stack([labels[i] for i in range(len(flag)) if flag[i]])
            if len(labels):
                optimizer.zero_grad()
                imgs = imgs.cuda()
                labels = labels.cuda()
                y_pred = model(imgs)
                loss = criterion(y_pred, labels)
                pbar.set_description(f"Loss: {loss.item():.6f}")
                loss.backward()
                optimizer.step()


        model_path = os.path.join(path, f"ConvNeXt_full_epoch_{epoch + 10}.pth")
        torch.save(model.state_dict(), model_path)
        print(f"Model saved to: {model_path}")

    print("Training complete.")


train_full_data(model, criterion, optimizer, train_loader, num_epoch=20, path="./")
```
