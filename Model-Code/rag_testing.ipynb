```python
import pandas as pd
import openai
import faiss
from sklearn.preprocessing import normalize
from dotenv import load_dotenv
import os

load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")


csv_file = "/content/language-of-flowers.csv"
df = pd.read_csv(csv_file)
df['Color'] = df['Color'].fillna('Unknown')
embedding_model = "text-embedding-ada-002"

def get_embedding(text):
    try:
        response = openai.Embedding.create(
            input=text,
            model=embedding_model
        )
        return response['data'][0]['embedding']
    except Exception as e:
        print(f"Error generating embedding: {e}")
        return None


print("Generating embeddings...")
df['embedding'] = df.apply(
    lambda row: get_embedding(f"{row['Flower']} {row['Color']} {row['Meaning']}"), axis=1
)


df = df[df['embedding'].notna()]

embeddings = list(df['embedding'])
normalized_embeddings = normalize(embeddings)

dimension = len(embeddings[0])
index = faiss.IndexFlatL2(dimension)
index.add(normalized_embeddings)


def query_rag_model(query, top_k=3):
    query_embedding = normalize([get_embedding(query)])
    distances, indices = index.search(query_embedding, top_k)
    results = df.iloc[indices[0]].drop(columns=['embedding'])
    return results


query = input("Enter a flower name, color, or meaning to search: ")
results = query_rag_model(query)


print("\nTop Matches:")
print(results)
```
